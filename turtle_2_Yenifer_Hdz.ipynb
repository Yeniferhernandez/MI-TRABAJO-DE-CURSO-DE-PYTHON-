```python
grados = 360
# lados = int(input('cuantos lados quieres tu figura: '))
# longitud = int(input('longitud de cada lado: '))
lados = 7
longitud = 60

lados_grados = grados / lados

initializeTurtle(initial_speed=10)

pencolor("purple")

for i in range(lados):
  pencolor("pink")
  forward(100)
  left(lados_grados)
  for i in range(lados):
    pencolor("yellow")
    forward(60)
    right(lados_grados)
    for i in range(lados):
      pencolor("white")
      forward(30)
      left(lados_grados)
```
